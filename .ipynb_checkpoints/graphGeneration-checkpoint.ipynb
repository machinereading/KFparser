```python
def get_text(sent, begin, end):
    text_list = []
    for token in sent:
        tokid = token[0]
        if tokid >= begin and tokid <= end:
            text_list.append(token[1])
    text = ' '.join(text_list)
    return text

def conll2kbox(sent_list):
    result = []
    for sent in sent_list:
        target, frame, fes = None, None, []
        for token in sent:
            if token[12] != '_':
                frame = 'frame:'+token[13]
                frame_id = token[0]
                break
        for token in sent:
            if token[14] != 'O':
                fes.append(token[14])
        fe_items = list(set(fes))
        fes = []
        for i in fe_items:
            span = []
            for token in sent:
                if i == token[14]:
                    span.append(token[0])
            if len(span) > 0:
                begin = span[0]
                end = span[-1]
#                 text = get_text(sent, begin, end)
                tu = (i, begin, end)
                fes.append(tu)
        
def conll2graph(sent_list):
    triples = []
    for sent in sent_list:
        target, frame, fes= None, None, []
        for token in sent:
            if token[12] != '_':
                lu = token[12]
                frame = token[13]
                break
        for token in sent:
            if token[14] != 'O':
                fes.append(token[14])
        # BIO tagging 고려 필요
        fe_items = list(set(fes))
        fes = []
        for i in fe_items:
            span = []
            for token in sent:
                if i == token[14]:
                    span.append(token[0])
            if len(span) > 0:
                begin = span[0]
                end = span[-1]
                text = get_text(sent, begin, end)
                tu = (i, text)
                fes.append(tu)
        
        triple = ('frame:'+frame, 'frdf:lu', lu)
        triples.append(triple)
        for i in fes:
            triple = ('frame:'+frame, 'fe:'+i[0].lower(), i[1])
            triples.append(triple)
    return triples
```
